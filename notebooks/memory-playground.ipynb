# Memory Playground

Experiment with short-term and long-term memory systems.
Tests memory storage, recall, and search capabilities.

```python
import requests
import json

API_URL = "http://localhost:4050"
API_KEY = ""

headers = {"Content-Type": "application/json"}
if API_KEY:
    headers["Authorization"] = f"Bearer {API_KEY}"

print("Memory Playground Ready")
print(f"API: {API_URL}")
```

```python
# Test knowledge base query (existing tool)
agent_id = "ceo"
prompt = "Search the knowledge base for information about our company strategy."

r = requests.post(
    f"{API_URL}/api/agents/{agent_id}/run",
    headers=headers,
    json={"input": {"prompt": prompt}}
)
print(json.dumps(r.json(), indent=2))
```

## Memory Architecture

- **Short-term**: In-memory buffer per agent session (bounded, auto-evicts)
- **Long-term**: Weaviate vector store with embedding-based recall
- **Tools**: `memory-recall` (search) and `memory-store` (persist)
