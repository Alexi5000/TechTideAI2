# Tool Calling Playground

Test tool execution interactively via the API.
Explore how agents use their tools to accomplish tasks.

```python
import requests
import json

API_URL = "http://localhost:4050"
API_KEY = ""

headers = {"Content-Type": "application/json"}
if API_KEY:
    headers["Authorization"] = f"Bearer {API_KEY}"

print("Tool Calling Playground Ready")
```

```python
# List all agents and their tools
r = requests.get(f"{API_URL}/api/agents", headers=headers)
agents = r.json()

print("Agent Tools Map:")
for agent in agents[:15]:
    tools = agent.get('tools', [])
    print(f"  {agent['id']:20s} -> {', '.join(tools)}")
```

```python
# Trigger a tool-heavy agent run
agent_id = "ceo"
prompt = "Use the org-kpi-dashboard to review our current performance metrics."

r = requests.post(
    f"{API_URL}/api/agents/{agent_id}/run",
    headers=headers,
    json={"input": {"prompt": prompt}}
)
run = r.json()
print(f"Run ID: {run.get('id')}")
print(f"Status: {run.get('status')}")
```
