# Prompt Engineering Playground

Interactive environment for iterating on agent prompt templates.
Uses the TechTideAI API to test prompts against live agents.

```python
import requests
import json

API_URL = "http://localhost:4050"
API_KEY = ""  # Leave empty for dev mode

headers = {"Content-Type": "application/json"}
if API_KEY:
    headers["Authorization"] = f"Bearer {API_KEY}"

# Verify connection
r = requests.get(f"{API_URL}/health")
print(f"API Status: {r.json()}")
```

```python
# List available agents
r = requests.get(f"{API_URL}/api/agents", headers=headers)
agents = r.json()
for agent in agents[:10]:
    print(f"  {agent['id']:20s} | {agent['name']:20s} | {agent['tier']}")
```

```python
# Test a prompt against an agent
agent_id = "ceo"
prompt = "What are our top 3 strategic priorities for this quarter?"

r = requests.post(
    f"{API_URL}/api/agents/{agent_id}/run",
    headers=headers,
    json={"input": {"prompt": prompt}}
)
print(json.dumps(r.json(), indent=2))
```
